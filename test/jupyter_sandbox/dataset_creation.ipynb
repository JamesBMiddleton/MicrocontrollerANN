```python
import numpy as np
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
X, y = make_regression(750, 2, noise=0.01)
```

```python
X = X.astype("float32")
y = y.astype("float32")

scaler = MinMaxScaler()  # because output is sigmoid
y = scaler.fit_transform(y.reshape(-1, 1))
y = y.reshape(-1)

X_train, X_test, y_train, y_test = train_test_split(X, y)

scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)


print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
print(abs(X_train[:, 0]).min())
print(abs(X_train[:, 0]).max())
print(abs(X_train[:, 1]).min())
print(abs(X_train[:, 1]).max())
```

```python
np.savetxt("generated_regression/X_train.csv", X_train, delimiter = ',', newline = ',', fmt = "%f")
np.savetxt("generated_regression/X_test.csv", X_test, delimiter = ',', newline = ',', fmt = "%f")
np.savetxt("generated_regression/y_train.csv", y_train, delimiter = ',', newline = ',', fmt = "%f")
np.savetxt("generated_regression/y_test.csv", y_test, delimiter = ',', newline = ',', fmt = "%f")
```
